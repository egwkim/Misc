```python
import math
import matplotlib.pyplot as plt # pylint: disable=missing-module-source

a = 10
b = 10
c = 10

a_list = [10]
b_list = [10]
c_list = [10]

size = 100

for i in range(size):
    a = (a - math.floor(a * 0.3)) * 2
    a_list.append(a)
    b = b * 1.4
    b_list.append(b)
    c = (c - math.ceil(c * 0.3)) * 2
    c_list.append(c)

fig = plt.figure()
ax = fig.add_subplot()

ax.plot(range(size + 1), a_list, color='r')
ax.plot(range(size + 1), b_list, color='g')
ax.plot(range(size + 1), c_list, color='b')

ax.set_yscale('log')
plt.show()
```
