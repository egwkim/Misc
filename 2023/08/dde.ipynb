```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Delay logistic equation
# y' = a y (1 - y(t-1))
def dle(a: float, steps: int, max_steps: int, initial_values: np.ndarray):
    values = np.ndarray(max_steps)
    values[:steps] = initial_values
    for i in range(steps, max_steps):
        values[i] = values[i - 1]
        values[i] += a * values[i] * (1 - values[i - steps]) / steps
    return values
```

```python
a = 1
steps = 1000
max_steps = 20 * steps
values = dle(a, steps, max_steps, np.ones(steps) * 2)
plt.plot(np.arange(max_steps) / steps, values)
plt.show()
```

```python
a = 2
steps = 1000
max_steps = 20 * steps
values = dle(a, steps, max_steps, np.ones(steps) * 2)
plt.plot(np.arange(max_steps) / steps, values)
plt.show()
```

```python
a = np.pi / 2
steps = 1000
max_steps = 20 * steps
values = dle(a, steps, max_steps, np.ones(steps) * 2)
plt.plot(np.arange(max_steps) / steps, values)
plt.show()
```

Damping when $a=pi/2$ might be due to the error

```python
# Random inital condition
a = 2
steps = 1000
max_steps = 50 * steps
values = dle(a, steps, max_steps, np.random.random(steps) * 3)
plt.plot(np.arange(max_steps) / steps, values)
plt.show()
```

Initial condition does not affect the overall result.
